```python
import pandas as pd

df = pd.read_csv("soomgo_pros.csv")
df
```

```python
hired_top200 = (df.groupby('main_service')['hired_count']
          .sum()
          .sort_values(ascending=False)
          .head(200))

hired_keywords = hired_top200.index.tolist()
print(hired_keywords)
```

```python
gosu_top200 = (df.groupby('main_service')['id']
          .count()
          .sort_values(ascending=False)
          .head(200))

gosu_keywords = gosu_top200.index.tolist()
print(gosu_keywords)
```

```python
hired_set = set(hired_keywords)
gosu_set = set(gosu_keywords)
```

```python
both = hired_set & gosu_set        # 둘 다 포함
only_hired = hired_set - gosu_set  # 고용횟수 기준에만
only_gosu = gosu_set - hired_set   # 고수수 기준에만
```

```python
print(only_hired)
```

```python
print(only_gosu)
```
