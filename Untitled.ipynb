```python
import pickle
from collections import Counter
from kiwipiepy import Kiwi
import pandas as pd

reviews = pd.read_csv('soomgo_reviews_clean.csv')
reviews = reviews.drop_duplicates(subset=['review_id'])

kiwi = Kiwi()
stopwords = ['없', '같', '있', '좋', '후', '때', '전', '일', '말', '이',
             '그', '수', '것', '더', '안', '못', '잘', '도', '만', '번']

def extract_words(text):
    if not isinstance(text, str):
        return []
    tokens = kiwi.analyze(text)[0][0]
    return [t.form for t in tokens if t.tag in ['NNG', 'VA', 'XR']]

low_r  = reviews[reviews['rating'] <= 3]['content'].dropna()
high_r = reviews[reviews['rating'] >= 4]['content'].dropna()

low_w, high_w = [], []
for text in low_r:  low_w.extend(extract_words(text))
for text in high_r: high_w.extend(extract_words(text))

low_f  = [w for w in low_w  if w not in stopwords and len(w) > 1]
high_f = [w for w in high_w if w not in stopwords and len(w) > 1]

with open('word_counters.pkl', 'wb') as f:
    pickle.dump((Counter(low_f), Counter(high_f)), f)

print("저장 완료!")
```
